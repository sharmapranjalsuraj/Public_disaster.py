# Natural Disasters all over the World

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing data in python:

```python
# Specify the file path and encoding
file_path = 'disaster.csv'
encoding = 'latin1'  # or 'ISO-8859-1' or 'cp1252'

# Read the CSV file into a pandas DataFrame
df = pd.read_csv(file_path, encoding=encoding)
```

# Cleaning the dataset.

```python
x = ['Classification Key', 'External IDs',  'Event Name', 'Associated Types', 'OFDA Response', 'Appeal', "AID Contribution ('000 US$)", 'River Basin', "Reconstruction Costs ('000 US$)",
       "Reconstruction Costs, Adjusted ('000 US$)",
       "Insured Damage ('000 US$)", "Insured Damage, Adjusted ('000 US$)",
       "Total Damage ('000 US$)", "Total Damage, Adjusted ('000 US$)", "CPI",
       "Admin Units", "Entry Date", "Last Update"]
df = df.drop(columns = x, axis = 1)
```

```python
df
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull()
```

Replace the NaN values with 0

```python
df.fillna(0, inplace = True)
```

#Change Data Type.

```python
df["Total Deaths"] = df["Total Deaths"].astype("int")
```

```python
df["Total Deaths"].dtype
```

# Exploratory Data Analysis:

Let's see the count of coparison between Natural Disaster and Technological Disaster.

```python
a = sns.countplot(x = "Disaster Group", data = df, width = 0.2)
for bars in a.containers:
    a.bar_label(bars)
conclusion_text = (
    "The analysis demonstrates a clear distinction in the occurrence of disasters, with a notably higher frequency "
    "of Natural Disasters compared to Technological Disasters."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=12, color='black', wrap=True)
plt.show(   )
```

Let's look at the deaths comparison between Natural and Technological Disaster.

```python
dis_group = df.groupby(["Disaster Group"], as_index = False)["Total Deaths"].sum()
dis_group = dis_group.sort_values("Total Deaths", ascending=False)
ax = sns.barplot(x = "Disaster Group", y = "Total Deaths", data = dis_group, width = 0.2)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
conclusion_text = (
    "The chart highlights a substantial discrepancy in the number of deaths caused by Natural and Technological Disasters. "
    "Natural Disasters stand out as the leading cause of fatalities, underscoring the severe impact these events can have on human "
    "populations."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=13, color='black', wrap=True)
plt.show()                                                                                    
```

Let's looked at count of Disasters in Regions.

```python
a = sns.countplot(data = df, x = "Region", hue = "Disaster Group")
for bars in a.containers:
    a.bar_label(bars)
conclusion_text = (
    "The chart provides insight into the distribution of disaster cases across continents, with a notable concentration in Asia. "
    "This observation suggests that Asia experiences a higher frequency of disaster events compared to other region."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=13, color='black', wrap=True)
plt.show()
```

Let's looked at deaths comarions in Top10 Countries.

```python
dis_group = df.groupby(["Country"], as_index = False)["Total Deaths"].sum()
dis_group = dis_group.sort_values("Total Deaths", ascending=False).head(10)
sns.set(rc={"figure.figsize": (15,5)})
ax = sns.barplot(x = "Country", y = "Total Deaths", data = dis_group, width = 0.4)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
conclusion_text = (
    "The chart illustrates that the highest number of death cases has occurred in China, followed by India."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=12, color='black', wrap=True)
plt.show()
```

Looking Forword let's see illustrating the top 10 causes of death by disaster types.

```python
dis_group = df.groupby(["Disaster Type"], as_index = False)["Total Deaths"].sum()
dis_group = dis_group.sort_values("Total Deaths", ascending=False).head(10)
sns.set(rc={"figure.figsize": (15,5)})
ax = sns.barplot(x = "Disaster Type", y = "Total Deaths", data = dis_group, width = 0.4)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
conclusion_text = (
    "The analysis highlights that the most significant causes of deaths are attributed to Drought, Epidemic, and Flood events. "
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=12, color='black', wrap=True)
plt.show()
```

Looking at number of injuries by Disaster types.

```python
dis_group = df.groupby(["Disaster Type"], as_index = False)["No. Injured"].sum()
dis_group = dis_group.sort_values("No. Injured", ascending=False).head(10)
sns.set(rc={"figure.figsize": (15,5)})
ax = sns.barplot(x = "Disaster Type", y = "No. Injured", data = dis_group, width = 0.4)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
conclusion_text = (
    "The analysis reveals that the most significant causes of injuries are associated with Earthquake and Epidemic events."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=12, color='black', wrap=True)
plt.show()
```

Let's see the percentage % of Deaths in Region.

```python
his = df.groupby(["Region"], as_index = False)["Total Deaths"].sum()
his = his.sort_values("Total Deaths", ascending=False)
z = [0.0,0.2,0.2,0.2,0.0]
plt.title("% of deaths in Region")
c = "y", "orange", "r", "blue", "green"
plt.pie("Total Deaths", labels = "Region", data = his, labeldistance = 1.2, textprops = {"fontsize" : 12}, 
        autopct = "%0.1f%%", shadow = True, radius = 1.2, colors = c, explode = z)
plt.legend()
conclusion_text = (
    "The analysis of death cases by continent reveals a notable distribution, with the majority of cases occurring in Asia "
    "(79.5%). Europe follows with 12.9%, Africa with 4.7%, Americas with 2.7%, and Oceania with a minimal 0.1%. This data underscores "
    "the significance of understanding regional disparities in disaster impacts and calls for targeted interventions and preparedness "
    "efforts, particularly in regions with higher vulnerability."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=13, color='black', wrap=True)
plt.show()
```

Looked at sum of deaths in Subregion.

```python
dis_group = df.groupby(["Subregion"], as_index = False)["Total Deaths"].sum()
dis_group = dis_group.sort_values("Total Deaths", ascending=False).head(10)
sns.set(rc={"figure.figsize": (20,5)})
ax = sns.barplot(x = "Subregion", y = "Total Deaths", data = dis_group, width = 0.4)
for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
conclusion_text = (
    "The analysis highlights that the regions with the highest number of deaths are Eastern Asia and Southern Asia. "
    "This observation underscores the disproportionate impact of disasters on these regions and emphasizes the need for targeted "
    "interventions, robust disaster management strategies, and community resilience-building initiatives in Eastern Asia and Southern Asia."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=13, color='black', wrap=True)
plt.show()
```

Let's see the declaration counts of the Disasters.

```python
a = sns.countplot(x = "Declaration", data = df, width = 0.2)
for bars in a.containers:
    a.bar_label(bars)
conclusion_text = (
    "This observation suggests that there might be instances where communities are not aware or prepared for an impending disaster."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=13, color='black', wrap=True)
plt.show()
```

Let's looked at top 10 Declaration disaster tyoe.

```python
top_disasters = df['Disaster Type'].value_counts().nlargest(10).index
df_top10 = df[df['Disaster Type'].isin(top_disasters)]
a = sns.countplot(data=df_top10, x="Disaster Type", hue="Declaration")
sns.set(rc={"figure.figsize": (15, 5)})
for bars in a.containers:
    a.bar_label(bars)
conclusion_text = (
    "The analysis reveals that the top 10 disaster types with the most declarations include "
    "instances where communities may not have been adequately prepared. The notably high declaration "
    "counts emphasize the need for enhanced awareness and preparedness measures. Further proactive measures "
    "can be taken to educate communities about potential disasters and encourage preparedness initiatives."
)
plt.annotate(conclusion_text,
             xy=(0.5, -0.2), xycoords='axes fraction',
             ha='center', va='center', fontsize=12, color='black', wrap=True)
plt.show()
```
